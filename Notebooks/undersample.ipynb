```python

import pandas as pd
import os

import seaborn as sns
```

```python
# --- Configuration ---
DATA_DIR = '../data/csv_data/'
SAMPLE_DIR = '../data/sample/'
if not os.path.exists(SAMPLE_DIR):
    os.makedirs(SAMPLE_DIR)

ACTOR_FILE = 'sub_sample.csv'  # CSV with a list of Actor IDs
PLAYER_FILE = '(after) Player information features.csv'
ACTION_FILE = '(after) Player actions features.csv'
SOCIAL_FILE = '(after) Social interaction diversity features.csv'
NETWORK_FILE = '(after) Network measures features.csv'
GROUP_FILE = '(after) Group activities features.csv'
```

```python
actor_df = pd.read_csv(os.path.join(DATA_DIR, ACTOR_FILE))
actor_ids = set(actor_df['Actor'].tolist())
print(f"Actors to verify: {actor_ids}")
```

```python
def extract_actor_data(actor_ids, data_df, filename):
    """Extracts data for specified actors and saves it to a CSV file."""
    actor_data = data_df[data_df['Actor'].isin(actor_ids)]
    if not actor_data.empty:
        filepath = os.path.join(SAMPLE_DIR, filename)
        actor_data.to_csv(filepath, index=False)
        print(f"Extracted data saved to: {filepath}")
    else:
        print(f"No data found for the specified actors in {filename}")
```

```python
try:
    # Read the CSV file with the list of actors
    actor_df = pd.read_csv(os.path.join(DATA_DIR, ACTOR_FILE))
    actor_ids = actor_df['Actor'].tolist()
    print(f"Actors to extract: {actor_ids}")

    # Load dataframes
    player_df = pd.read_csv(os.path.join(DATA_DIR, PLAYER_FILE))
    action_df = pd.read_csv(os.path.join(DATA_DIR, ACTION_FILE))
    social_df = pd.read_csv(os.path.join(DATA_DIR, SOCIAL_FILE))
    network_df = pd.read_csv(os.path.join(DATA_DIR, NETWORK_FILE))
    group_df = pd.read_csv(os.path.join(DATA_DIR, GROUP_FILE))

    # Extract data for each actor
    extract_actor_data(actor_ids, player_df, 'sample_player_data.csv')
    extract_actor_data(actor_ids, action_df, 'sample_action_data.csv')
    extract_actor_data(actor_ids, social_df, 'sample_social_data.csv')
    extract_actor_data(actor_ids, network_df, 'sample_network_data.csv')
    extract_actor_data(actor_ids, group_df, 'sample_group_data.csv')

except FileNotFoundError as e:
    print(f"File not found: {e}")
except KeyError as e:
    print(f"Key Error: {e}. Please ensure the 'Actor' column exists in all files.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
